```python
import networkx as nx
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
D = nx.DiGraph()
D.add_nodes_from([1,2,3,4,5,6])
D.add_edges_from([(1,2),(2,3),(3,2),(2,4),(4,1),(3,1),(6,3),(6,1)])
nx.draw(D, with_labels=True, node_color='red', font_color='white')
```

```python

%matplotlib inline

G = nx.Graph()

G.add_nodes_from([1,2,3,4,5,6])

G.add_edges_from([(1,2),(2,3),(1,3),(1,4),(2,4),(6,1),(6,3)])

nx.draw(G, with_labels=True,node_color='red', font_color='white')
```

```python
degree_sequence = [G.degree(n) for n in G.nodes]
degree_counts = Counter(degree_sequence)
min_degree, max_degree = min(degree_counts.keys()), max(degree_counts.keys())
plt.xlabel("Degree", fontsize=10,color='green')
plt.ylabel("Number of Nodes", fontsize=10,color='green')
plot_x = list(range(min_degree, max_degree + 1))
plot_y = [degree_counts.get(x, 0) for x in plot_x]
plt.bar(plot_x, plot_y)
```

```python
def plot_Degree_dist(G):  
    degrees = G.degree()
    degrees = dict(degrees)
    values = sorted(set(degrees.values()))
    print(values)
    histo = [list(degrees.values()).count(x) for x in values]
    P_k = [x / G.order() for x in histo]
    print(len(P_k))
    
    plt.figure()
    plt.bar(values, P_k)
    plt.xlabel("k",fontsize=10,color='green')
    plt.ylabel("p(k)", fontsize=10,color='green')
    plt.title("Degree Distribution", fontsize=10,color='green')
    
    plt.show()
    plt.figure()
    plt.grid(False)
    plt.loglog(values, P_k, "bo")
    plt.xlabel("k", fontsize=10)
    plt.ylabel("log p(k)", fontsize=10,color='green')
    plt.title("log Degree Distribution")
    plt.show()
    plt.show()
plot_degree_dist(G)
```
